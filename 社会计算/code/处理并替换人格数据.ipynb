```python
import pandas as pd
```

```python
txt_path = "问题正负的映射.txt"
# 使用pandas read_csv函数读取txt文件，指定分隔符为'|'，并跳过第一行（表头行下方的分隔符行）
df = pd.read_csv(txt_path, sep='|', skiprows=1)

# 去掉可能的首尾空格
df.columns = [col.strip() for col in df.columns]
df = df.applymap(lambda x: x.strip() if isinstance(x, str) else x)
df = df.drop(columns=['Unnamed: 0','Unnamed: 4',':----.1'])

# 重命名列
df = df.rename(columns={':---': 'label', ':----': 'phrased'})

df
```

```python
data_path = 'data-final.csv'

df1 = pd.read_csv(data_path, delimiter='\t')
print(df1.columns)
```

```python
# 创建映射函数
def map_values(row, label_phrased_dict):
    for label, phrased in label_phrased_dict.items():
        if phrased == 'N':
            row[label] = 6 - row[label]
    return row

# 创建 label-phrased 映射字典
label_phrased_dict = dict(zip(df['label'], df['phrased']))

# 对 df1 应用映射
df2 = df1.apply(lambda row: map_values(row, label_phrased_dict), axis=1)

df2
```

```python
df2.to_csv('data-final_修正版.csv')
```
